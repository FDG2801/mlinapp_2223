# Names

##  Create the 3 DataFrames based on the following raw data

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

## Assign each to a variable called data1, data2, data3

```python
data1 = pd.DataFrame(raw_data_1, columns = ['subject_id','first_name','last_name'])
data2 = pd.DataFrame(raw_data_2, columns = ['subject_id','first_name','last_name'])
data3 = pd.DataFrame(raw_data_3, columns = ['subject_id', 'test_id'])
```

## Join the two dataframes along rows and assign all_data

```python
all_data = pd.concat([data1,data2])
all_data
```

## Join the two dataframes along columns and assing to all_data_col

```python
all_data_col = pd.concat([data1,data2], axis=1)
all_data_col
```

## Merge all_data and data3 along the subject_id value. Is there nay NaN value?

```python
pd.merge(all_data, data3, on='subject_id')
#no NaN value
```

## Merge only the data that has the same 'subject_id' on both data1 and data2

```python
print(data1)
print(data2)
pd.merge(data1,data2, on='subject_id', how = 'inner')
```

## Merge all values in data1 and data2, with matching records from both sides where available.

```python
pd.merge(data1,data2, on='subject_id', how = 'outer')
```

# Drinks

## Read and save data from datasets/drinks.txt

```python
#there is no drinks txt... only drinks.csv
FILEPATH = "/content/drive/MyDrive/mlinapp/basics-data/drinks.csv"
drinks = pd.read_csv(FILEPATH)
print(drinks)
print(drinks.to_string())
```

## Which continent drinks more beer on average?

```python
drinks.groupby('continent').beer_servings.mean().sort_values(ascending = False)
```

## For each continent print the statistics for wine consumption.

```python
drinks.groupby('continent').wine_servings.describe()
```

##  Print the mean, min and max values for spirit consumption per continent.

```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```
